#### Dataset  Description: 

    The dataset is of the shows that runs daily on our TVs. The datapoints correspond to each TV show and the columns correspond to variaous attributes of the TV show. This data will be useful for the Producer and Channel broadcasting the show for their economic growth. Thats means, these members can find out before hand or after just launching the show that how many maximum viewers the show could have (based on its features) and whether the show would be hit/popular or run successfully. So, this is collectively termed to 'watch' feature (in our case), which is our dependent/response variable. And the features that determine this would be TV show's genre, language, various ratings (TRP in our case), duration of the show, which here are considered as independent/predictor variables.
    
    We, have one more attribute named 'time' i.e, what time of the day, the show is aired. Note: this attribute is not used here. But, we can use it in a way that finds out what/which current shows, how many people are watching and find out by analysis if the viewers could switch to new show when aired irrespective of the viewers interest(of the current show) and if the new show timings collides with current show timings.
    
    

```python
# 1.Import the required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 2.Read the dataset into a variable

data = pd.read_csv("C:/Users/Sai/Desktop/shows.csv",encoding='cp1252')
data
```

### Feature Engineering

```python
# 3.Encoding values of some features
data['genre'] = data['genre'].map({'comedy':0, 'reality':1, 'drama':2, 'historic':3, 'fantasy':4, 'game':5})
data['language'] = data['language'].map({'hindi':0, 'telugu': 1, 'tamil': 2})
```

```python
data
```

```python
# 4.Creating a empty list(feature)
a = []
a
```

```python
# 5.Changing the numeric variable to categorical with a particular condition( if value >= 70: 'yes' else 'no')
#   This condition acts somewhat as an external bias of perceptron model i.e, 'b'. We can change it to our requirement.

for c in data['watch_percent']:
    if c >= 70:
        a.append("yes")
    else:
        a.append("no")   
```

```python
# 6.Printing the newly obtained feature
a
```

```python
# 7.Appending the obtained feature to our dataset
data['watch'] = a
data

# Therefore now our dependent varaible is 'watch'
```

```python
# 8.Encoding values of 'watch' features
data['watch'] = data['watch'].map({'yes':1, 'no':0})
data
```

```python
# 9.Retaining features which contribute to the model-output is the 'watch' attribute

df = data[['genre', 'language', 'duration', 'episodes', 'trp_rating', 'watch_percent', 'watch']]
df
```

```python
# 10.Set the predictor(X) and Response(Y) variables.
# axis specify if its row:0 or columns:1

X = df.drop("watch", axis = 1) 
Y = df['watch']
```

### Train Test Split

```python
# 11.Splitting into training and test set

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state = 25)

print("Rows  Columns")
print("\n", X_train.shape,"\n", Y_train.shape,"\n", X_test.shape,"\n", Y_test.shape)
```

###  Classification using Logistic Regression Classifier

```python
# 12.Calling the logistic regression model
from sklearn.linear_model import LogisticRegression
logit = LogisticRegression()
```

```python
# 13.fit() gives the perfect values for the arg of logistic regression
logit.fit(X_train, Y_train)
```

```python
# 14.Calling the logistic regression model again with 200 iterations

logit = LogisticRegression(max_iter=200)
```

```python
# 15.fit() gives the perfect values for the arg of logistic regression
logit.fit(X_train, Y_train)
```

```python
# 16.Compute the prediction or y hat

Y_pred = logit.predict(X_test)

# Printing the predicted values (y-hat)
print(Y_pred)
```

```python
# 17.Printing the Y's test values (y)
print(Y_test)
```

### Accuracy Metrics

```python
# 18.Confusion Matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns

confusion_matrix = confusion_matrix(Y_test, Y_pred)

# below stmts are used for annotations of confusion-matrix visualizing using heatmap

# names/labels of each cell
group_names = ['True Neg', 'False Pos', 'False Neg', 'True Pos']

# count of each cell i.e the total values under each label
group_counts = ["{0:0.0f}".format(value) for value in confusion_matrix.flatten()]

# corresponding %
group_percentages = ["{0:.2%}".format(value) for value in confusion_matrix.flatten()/np.sum(confusion_matrix)]

labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in zip(group_names,group_counts,group_percentages)]

labels = np.asarray(labels).reshape(2,2)

annot_kws={'fontsize':20}

sns.heatmap(confusion_matrix, annot=labels, fmt='', cmap='Blues', annot_kws=annot_kws)


#A heat map is a data visualization technique that shows magnitude of a phenomenon as color(color-coding to represent different values) in two dimensions.
# annot is used to add the required labels(True Neg, 1, 25.00%) to each heatmap cell
# fmt: annot will only help to add numeric value on python heatmap cell but fmt allows to add string values on the cell
# cmap: to change color of heatmap.
# annot_kws: used to style and format annot(text in cell) 
```

```python
# 19.Accuracy Score OF LOGISTIC REGRESSION CLASSIFIER
from sklearn.metrics import accuracy_score
print('Accuracy Score')
accuracy_score(Y_test, Y_pred)*100
```

```python
# 20.Classification Report
from sklearn.metrics import classification_report
report = classification_report(Y_test, Y_pred)
print(report)
```
